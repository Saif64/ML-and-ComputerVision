```python
import cv2
import matplotlib.pyplot as plt
import  numpy as np
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```

## draw rectangle

```python
cv2.rectangle(blank_img, pt1= (300, 200), pt2=(450, 250), color=(0,255,0), thickness= 10)
```

```python
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1= (100, 400), pt2=(200, 300), color=(0,255,0), thickness= 10)
```

```python
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1= (200, 200), pt2=(100, 100), color=(0,9,255), thickness= 5)
```

```python
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```

## Draw circle

```python
cv2.circle(blank_img, center= (350,350), radius= 50, color=(255,0,0), thickness= 5)
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```

```python
# making the circle filled in
cv2.circle(blank_img, center= (50,50), radius= 30, color=(255,0,0), thickness= -1)
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```

## Drawing lines

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(0,0,255) , thickness= 5)
fig = plt.figure(figsize=(18, 10))
plt.imshow(blank_img)
```
